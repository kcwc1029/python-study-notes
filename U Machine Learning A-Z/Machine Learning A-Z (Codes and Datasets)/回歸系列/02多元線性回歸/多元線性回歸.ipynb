# 多元線性回歸

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('50_Startups.csv')
column_names = dataset.columns.tolist()
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values

print("資料名稱", column_names)
print(X)
print(y) # 利潤 
```

## 資料進行熱處理

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')
X = np.array(ct.fit_transform(X))

print(X)
```

```python
# 拆分資料
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# 訓練模型

from sklearn.linear_model import LinearRegression
# 創建線性回歸模型的實例
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# 預測資料
y_pred = regressor.predict(X_test)

# 設定 numpy 的打印選項，設定為兩位小數
np.set_printoptions(precision=2)

# 打印預測結果和實際值
# - np.concatenate((array1, array2), axis=1) 將兩個二維數組沿著列方向（axis=1）拼接起來
# - y_pred.reshape(len(y_pred),1) 和 y_test.reshape(len(y_test),1) 將一維數組轉換為列向量
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)), axis=1))
```

```python
# 可視化
import matplotlib.pyplot as plt

# 設定圖形大小
plt.figure(figsize=(10, 6))

# 繪製預測值和實際值的散點圖
plt.scatter(y_test, y_pred, color='blue', label='Predicted vs. Actual')

# 繪製一條直線表示理想狀況（預測值等於實際值）
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--', lw=2, label='Ideal')

# 添加圖形標籤和標題
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Predicted vs. Actual')
plt.legend()

# 顯示圖形
plt.show()
```
