## 深度學習中最常見的錯誤之一（形狀錯誤）

```python
# Shapes need to be in the right way
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]], dtype=torch.float32)

tensor_B = torch.tensor([[7, 10],
                         [8, 11],
                         [9, 12]], dtype=torch.float32)

torch.matmul(tensor_A, tensor_B) # (this will error)
```

他這邊有寫，因為你拿3x2 and 3x2座矩陣乘法導致錯誤。

修正方式就是要將矩陣做轉至。
- `torch.transpose(input, dim0, dim1)` - 其中 input 是要轉置的所需張量， dim0 和 dim1 是要交換的維度。
- `tensor.T` - tensor 是要轉置的所需張量。

```python
# View tensor_A and tensor_B.T
print(tensor_A)
print(tensor_B.T)
```

```python
print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}\n")
print(f"New shapes: tensor_A = {tensor_A.shape} (same as above), tensor_B.T = {tensor_B.T.shape}\n")
print(f"Multiplying: {tensor_A.shape} * {tensor_B.T.shape} <- inner dimensions match\n")
print("Output:\n")
output = torch.matmul(tensor_A, tensor_B.T)
print(output)
print(f"\nOutput shape: {output.shape}")
```

而為什麼要做矩陣乘法呢？

神經網路充滿了矩陣乘法和點積。

像是 `torch.nn.Linear()` 模組（我們稍後將看到它的實際應用），也稱為前饋層或全連接層，實現了輸入 x 矩陣和權重矩陣之間的矩陣乘法。



$$
y = x \cdot A^T + b
$$

## 重塑、堆疊、擠壓和取消擠壓

＄ 很多時候，您需要重塑或更改張量的維度，而無需實際更改其中的值。

- torch.reshape(input, shape)：
- Tensor.view(shape)：返回不同 shape 但與原始張量共用相同數據的原始張量的檢視。
- torch.stack(tensors, dim=0)：沿新維度 （ dim ） 連接的 tensors 序列 tensors 必須大小相同。
- torch.squeeze(input)：擠壓 input 以刪除所有具有值 1 的尺寸。
- torch.unsqueeze(input, dim)：返回 input 的維度值 1 為 。
- torch.permute(input, dims)：傳回原始 input 檢視，其尺寸排列（重新排列）為 dims 。

```python
# Create a tensor
import torch

x = torch.arange(1., 8.)
x, x.shape
```

```python
# torch.reshape() 將一維做成二維
x_reshaped = x.reshape(1, 7)
x_reshaped, x_reshaped.shape
```

```python
# 使用torch.view()更改檢視
z = x.view(1, 7)
z, z.shape
```

```python
# 使用 torch.stack()將新張量堆疊在自身之上五次
# 嘗試更改dim=0，dim=1
print(x.shape)
x_stacked = torch.stack([x, x, x, x], dim=0)
print(x_stacked)
print(x_stacked.shape)

x_stacked = torch.stack([x, x, x, x], dim=1)
print(x_stacked)
print(x_stacked.shape)
```

假如你將 x 沿著 dim=0 堆疊，這意味著你想要將 x 疊起來形成一個新的張量，新的張量的形狀中第一個維度將會是堆疊的次數。舉例來說，如果你把 x 疊四次，那麼結果的形狀會是 (4, 3)，意思是有四個元素，每個元素都是一個三維的向量。

而如果你將 x 沿著 dim=1 堆疊，這意味著你想要將 x 的元素在同一個維度上串接在一起，也就是把它們排成一行。舉例來說，如果你把 x 疊四次，那麼結果的形狀會是 (3, 4)，意思是有三個元素，每個元素都是一個包含四個值的向量。

使用 torch.squeeze()壓縮維度為1的維度。

```python
print(f"Previous tensor: {x_reshaped}")
print(f"Previous shape: {x_reshaped.shape}")

# Remove extra dimension from x_reshaped
x_squeezed = x_reshaped.squeeze()
print(f"\nNew tensor: {x_squeezed}")
print(f"New shape: {x_squeezed.shape}")
```

相反 `torch.squeeze()` ，您可以使用 `torch.unsqueeze()` 在特定索引處添加維度值 1。

```python
print(f"Previous tensor: {x_squeezed}")
print(f"Previous shape: {x_squeezed.shape}")

## Add an extra dimension with unsqueeze
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(f"\nNew tensor: {x_unsqueezed}")
print(f"New shape: {x_unsqueezed.shape}")
```

```python
# 使用 torch.permute(input, dims) 重新排列軸值的順序，其中 input 將轉換為具有 new dims 的檢視。
# Create tensor with specific shape
x_original = torch.rand(size=(224, 224, 3))

# Permute the original tensor to rearrange the axis order
x_permuted = x_original.permute(2, 0, 1) # shifts axis 0->1, 1->2, 2->0

print(f"Previous shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}")
```

## PyTorch 張量和 NumPy

- torch.from_numpy(ndarray) - NumPy 陣列 -> PyTorch 張量。
- torch.Tensor.numpy() - PyTorch 張量 -> NumPy 陣列。

```python
# NumPy array to tensor
import torch
import numpy as np
array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array)
array, tensor
```

```python
# Tensor to NumPy array
tensor = torch.ones(7) # create a tensor of ones with dtype=float32
numpy_tensor = tensor.numpy() # will be dtype=float32 unless changed
tensor, numpy_tensor
```
